```python
import pandas as pd
from sklearn.model_selection import train_test_split

# 1. Membaca data film dari Google Drive
path_file = '/content/drive/MyDrive/MovieData-2018-2020.csv'
data = pd.read_csv(path_file)

# 2. Menghilangkan genre yang hanya berisi 1 judul film agar stratified sampling lancar
filter_data = data.groupby('genre').filter(lambda x: len(x) > 1)

print("Data berhasil dimuat dan dibersihkan!")
```

```python
# Melakukan Stratified Sampling menggunakan NIM
_, stratified_sample = train_test_split(
    filter_data,
    test_size=0.2,
    stratify=filter_data['genre'],
    random_state=47
)

# Mengambil 30 data teratas dari hasil sampling
sample = stratified_sample.head(30)
data_budget = sample['budget']

# Menampilkan 30 data teratas sampel
print(" Stratified Sampling (30 data teratas NIM 047) ")
sample
```

```python
# Menghitung statistik deskriptif dari data budget sampel
print(data_budget.describe())
```

```python
import scipy.stats as stats
import math

# Mengambil statistik dari data budget sampel secara otomatis
x_bar = data_budget.mean()
s = data_budget.std()
n = 30

# Menentukan Tingkat Kepercayaan (Confidence Level)
confidence_level = 0.90

# Logika Perhitungan T-Distribution
alpha = (1 - confidence_level)
df = n - 1
standard_error = s / math.sqrt(n)
critical_t = stats.t.ppf(1 - alpha/2, df)
margin_of_error = critical_t * standard_error

# Menghitung Batas Bawah dan Batas Atas Interval Kepercayaan
lower_bound = x_bar - margin_of_error
upper_bound = x_bar + margin_of_error

# Menampilkan Hasil Akhir Tugas
print("HASIL PERHITUNGAN ESTIMASI POPULASI ")
print("Critical t-value   : {:.2f}".format(critical_t))
print("Margin of Error    : {:,.2f}".format(margin_of_error))
print("Confidence Interval: [{:,.2f} , {:,.2f}]".format(lower_bound, upper_bound))
print("\n KESIMPULAN ")
print("The {:.1%} confidence interval for the population mean is:".format(confidence_level))
print("between {:,.2f} and {:,.2f}".format(lower_bound, upper_bound))
```
